```python
class DatosGral:
  def __init__(self):
    self._NonCien = "" #Atributo protegido
    self.Almacenar = []
    self._Ubicación = ""
    self._Clasificación = ""
    self.PelExtinción = ""
    self.NomVulgar = ""
    self.Alimentación = ""
    self.Población = 0 #Recibe un entero
  def Imprimir(self):
    print("Nombre ciéntifico: ", self._NonCien)
    print("Ubicación: ",self._Ubicación)
    print("Peligro de extinción: ",self.PelExtinción)
    print("Clasificación: ", self._Clasificación)
  def Ingresar_Datos(self,reinado):
    self.info = []
    self._Clasificación = reinado + str(len(self.Almacenar)).zfill(4)
    self._NonCien = input("Ingresa el nombre científico: ")
    self._Ubicación = input("Ingresa el lugar geográfico: ")
    self.PelExtinción = input("¿Esta en peligro de extinción? Sí/No: ")
    if self.PelExtinción.upper() ==  "SI":
      self.PelExtinción = True
    elif self.PelExtinción.upper() ==  "NO":
      self.PelExtinción = False
    else:
      self.PelExtinción = False
      print("Tu respuesta no coincide con ninguna de las opciones")
    self._NomVulgar = input("Ingresa el nombre vulgar: ")
    self.Alimentación = input("Ingresa los alimentos: ")
    self.Población = int(input("Ingresa el numero de población: "))
    self.info.append(self._Clasificación)
    self.info.append(self._NonCien)
    self.info.append(self._Ubicación)
    self.info.append(self.PelExtinción)
    self.info.append(self._NomVulgar)
    self.info.append(self.Alimentación)
    self.info.append(self.Población)
    self.Almacenar.append(self.info)
  def Eliminar_Registro (self, clasi):
    self.buscar = clasi
    for i in range(len(self.Almacenar)):
      if self.Almacenar[i][0] == self.buscar:
        self.Almacenar.pop([i][0])
        break
  def Actualizar_Registro (self,clasi,ayuda):
    self.buscar = clasi
    print("A continuación se muestran los elementos que puedes actualizar\n 1- Nombre Ciéntifico \n 2- Ubicación \n 3- Peligro de Extención \n 4- Nombre vulgar\n")
    print(" 5- Alimentación \n 6- Población")
    self.registro = int(input("¿Qué dato quieres actulizar?. Ingresa el número correspondiente al menú "))
    for i in range(len(ayuda.Almacenar)):
      if ayuda.Almacenar[i][0] == self.buscar:
        if self.registro == 1:
          ayuda.Almacenar[i][1] = input("Ingresa el nuevo nombre ciéntifico: ")
          break
        elif self.registro == 2:
          ayuda.Almacenar[i][2] = input("Ingresa la nueva ubicación: ")
          break
        elif self.registro == 3:
          ayuda.Almacenar[i][3] = input("¿Esta en peligro de extinción? SI/NO: ")
          if ayuda.Almacenar[i][3].upper() ==  "SI":
            ayuda.Almacenar[i][3] = True
          elif ayuda.Almacenar[i][3].upper() == "NO":
            ayuda.Almacenar[i][3] = False
          else:
            ayuda.Almacenar[i][3] = False
            print("Tu respuesta no coincide con ninguna de las opciones")
            break
          break
        elif self.registro == 4:
          ayuda.Almacenar[i][4] = input("Ingresa el nuevo nombre vulgar: ")
          break
        elif self.registro == 5:
          ayuda.Almacenar[i][5] = input("Ingresa el tipo de alimentación: ")
          break
        elif self.registro == 6:
          ayuda.Almacenar[i][6] = input("Ingresa la número de Población: ")
          break
        else:
          print("El número que ingresaste no correponde a nigún elemento del menú")
          break
      else:
        print("La clasifificación no fue encontrada")
        break
  def Consultar_Registro (self, clasi):
    self.buscar = clasi
    for i in range(len(self.Almacenar)):
      if self.Almacenar[i][0] == self.buscar:
        return self.Almacenar[i]
        break
class Animalia(DatosGral):
  def __init__(self):
    DatosGral.__init__(self)
class Plantae(DatosGral):
  def __init__(self):
    DatosGral.__init__(self)
class Monera(DatosGral):
  def __init__(self):
    DatosGral.__init__(self)
class Fungi(DatosGral):
  def __init__(self):
    DatosGral.__init__(self)
class Protista(DatosGral):
  def __init__(self):
    DatosGral.__init__(self)
Animal, Planta, Hongo, Bacteria, Proto = Animalia(), Plantae(), Fungi(), Monera(), Protista()
```

```python
while True:
  reinos = {1 : Animal, 2 : Planta, 3 : Hongo, 4 : Bacteria, 5 : Proto}
  q = {1 : "A", 2 : "P", 3 : "F", 4 : "M", 5 : "PR"}
  print("\nMenú")
  print(" 1- Nuevo Registro \n 2- Consulta \n 3- Actualizar registro \n 4- Eliminar \n 5- Salir")
  op = int(input("Elige cualquier opción del menú: "))
  if op >= 1 and op <= 5  :
    if  op == 5:
      break
    else:
      print(" 1- Animalia \n 2- Plantae \n 3- Fungi \n 4- Monera \n 5- Protista ")
      rey = int(input("¿Para que reino será la acción? "))
      if rey >= 1 and rey <= 5 :
        if op == 1:
          list(reinos.values())[rey-1].Ingresar_Datos(q[rey])
          break
        elif op == 2:
          buscando = input("Ingresa la clasificación del registro a consultar: ")
          encontrado = list(reinos.values())[rey-1].Consultar_Registro(buscando)
          if encontrado == None:
              print("El registro que buscas no fue encontrado")
              break
          else:
              print(encontrado)
              break
        elif op == 3:
          buscando = input("Ingresa la clasificación del registro actualizar: ")
          encontrado = list(reinos.values())[rey-1].Consultar_Registro(buscando)
          if encontrado == None:
              print("El registro que buscas no fue encontrado")
              break
          else:
              list(reinos.values())[rey-1].Actualizar_Registro(buscando,list(reinos.values())[rey-1])
              break
        elif op == 4:
          buscando = input("Ingresa la clasificación del registro a eliminar:")
          encontrado = list(reinos.values())[rey-1].Consultar_Registro(buscando)
          if encontrado == None:
              print("El registro que buscas no fue encontrado")
              break
          else:
              list(reinos.values())[rey-1].Eliminar_Registro(buscando)
              break
        else :print("La opción que ingresaste no esta en el menú")
  else: print("La opción que ingresaste no esta en el menú")
```
