```python
Letras = {"A": ["Á","Â","À","Ä"],"C": ["Ç"],"E": ["É", "Ê","È","Ë"],"I": ["Í","Ï","Î","Ì"],"O": ["Ó","Ô","Ö","Œ"],"U": ["Ú","Ü","Û","Ù"]}
Pais = ["Español","Inglés","Francés","Alemán","Italiano"]
Abecedario = {"A": [0,11.72,8.34,8.13,5.58,10.85], "B": [0,1.49,1.54,.93,1.96,1.05], "C": [0,.87,2.73,3.15,3.16,4.3], "D": [0,4.67,4.14,3.55,4.98,3.39], "E": [0,13.72,12.60,15.10,16.93,11.49], "F": [0,.69,2.03,.96,1.49,1.01], "G": [0,1,1.92,.973,.02,1.65], "H": [0,1.18,6.11,1.08,4.98,1.43], "I": [0,5.28,6.71,6.94,8.02,10.18], "J": [0,.52,.23,.71,.24,0],"K": [0,.11,.87,.16,1.32,0], "L": [0,5.24,4.24,5.68,3.60,5.7], "M": [0,3.08,2.53,3.23,2.55,2.87],"N": [0,6.83,6.80,6.42,10.53,7.02], "Ñ": [0,.17,0,0,0,0], "O": [0,8.44,7.70,5.27,2.24,9.97],"P": [0,2.89,1.66,3.03,.67,2.96], "Q": [0,1.11,.09,.89,.02,.45], "R": [0,6.41,5.68,6.43,6.89,6.19], "S": [0,7.20,6.11,7.91,6.42,5.48], "T": [0,4.60,9.37,7.11,5.79,6.97], "U": [0,4.55,2.85,6.05,3.83,3.16], "V": [0,1.05,1.06,1.83,.84,1.75], "W": [0,.04,2.34,.04,1.78,0], "X": [0,.14,.20,.42,.05,0], "Y": [0,1.09,2.04,.19,.05,0], "Z": [0,.47,.06,.21,1.21,.85]}
#Limpiar texto
def Texto(Archivo):
  nuevo = ""
  for n in Archivo.read():
    if n.isalpha():
      nuevo += n.upper()
  m,c = 0, list(nuevo)
  for s in nuevo:
      m += 1
      for i in Letras:
          for t in Letras[i]:
              if t == s:
                c[m-1] = i 
                break
  for z in Abecedario: #Porcentajes de cada letra
    Abecedario[z][0] = round(((c.count(z))/len(nuevo))*100,1)
  return c, nuevo
#¿Qué idioma es?
def Idiomas (Abecedario):
    p, suma, r, lista  = 0, 0, 5, []
    while r >= 1 and p < 27:
        for q in list(Abecedario.values()):
            suma += abs(list(Abecedario.values())[p][r]-q[0])
            p += 1
        lista.append(suma)
        p, suma = 0, 0
        r -= 1
    final = lista[::-1]
    for n in range(len(final)):
        if final[n] == min(final):
            return  Pais[n], n+1          
#Gráfica
def graficar (Abecedario,posicion,idi):
  import matplotlib.pyplot as plt
  import numpy as np
  fig , ax = plt.subplots(figsize = (28,10))
  ax.set_title("Porcentaje de aparición", fontweight ='bold', fontsize = 20)
  w,co, al = .40, [], []
  cota = np.arange(27)  #Porcentaje del texto
  cota2 = [x + w for x in cota]
  for u in list(Abecedario.values()):  #Porcentaje del idioma encontrado
    al.append(u[0])
  for t in list(Abecedario.values()):  #Porcentaje del idioma encontrado
    co.append(t[posicion])
  ax.bar(cota, al, color = "b", width = w , label = "Texto")
  ax.bar(cota2, co , color = "c", width = w , label = idi)
  [ax.annotate(h, xy = (g - .18, h + .1)) for g,h in zip(cota,al)]
  [ax.annotate(l, xy = (k - .2, l + .1)) for k,l in zip(cota2,co)]  
  plt.xlabel('Abecedario', fontweight ='bold', fontsize = 14)
  plt.ylabel('Porcentajes', fontweight ='bold', fontsize = 14)
  plt.xticks([r + .20 for r in range(27)],list(Abecedario.keys()))
  plt.legend(loc='upper left', bbox_to_anchor = (0.5, .9) )
  return plt.show()


#Main
while True:
    txt = input('Ingresa un archivo con extensión .txt. Verifica que se encuentre en la misma carpeta. : ')
    buscar = txt[::-1].find('.')
    posición = len(txt)-buscar-1
    if txt[posición:] != ".txt":
      print('El archivo no tiene extensión .txt. Intentalo de nuevo.')
      continue
    else:
      Doc = open(txt,'r+')
      print(f'\nEl texto que ingresaste es: \n\n {Doc.read()}')
      Doc.read()
      Doc.seek(0)
      [a,b] = Texto(Doc)
      [c,d] =  Idiomas(Abecedario)
      print(f'\n El texto que ingresaste probablemente esta en el idioma {c}.')
      graficar(Abecedario,d,c)
      Doc.close()
      break
```
